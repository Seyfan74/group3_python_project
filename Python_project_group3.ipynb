1)	Build a program that:

i.	Displays a list of snacks and drinks with item numbers and prices.

ii.	Ask the user to choose items by number in a loop.

iii. Keeps track of selected items and their prices.

iv.	Ends when the user types "done".

```python

# i.	Displays a list of snacks and drinks with item numbers and prices.

items= {
   1: ('chips', 1),
   2: ('pepsi', 1.5),
   3: ('donut', 2),
   4: ('coca-cola', 1.3),
   5: ('popcorn', 2.5)
}
selected_items =[]
total_price =0
print('print snacks in list =>', items)

#ii.	Ask the user to choose items by number in a loop.
for item_no, (name, price) in items. items():
    print(f'{item_no}. {name} ${price}')



print()
while True:
    user_choice= input('enter item number or done :')
    if user_choice.lower()=='done':
        break
    items_no= int(user_choice)
    
    if items_no in items:
        name, price = items[items_no]
        selected_items.append((name,price))
        total_price = total_price + price
        print(f' added {name}, -->${price}')
    else:
        print('invalid item number')

print('total_price',' =', '$',total_price)

print()
print('=='*10)
print(     'Receipt')
print('=='*10)
for name, price in selected_items:
    print(name, price)
print('=='*10)
print(f'Total_costs:   ${total_price}')
print('=='*10)
```

2)	Write a program that:
•	Has a predefined dictionary of groceries with prices.
•	Lets the user "add" items by typing their names.
•	For each valid item, asks for the quantity.
•	Keeps adding to the cart until the user types "checkout".
•	Displays a final bill: each item, quantity, subtotal, and total.

```python

# Has a predefined dictionary of groceries with prices.

# Dictionary containing grocery items and their prices.
groceries = {
    'milk': 1.5,
    'chocolate':2.5,
    'egg': 5,
    'water': 4.5,
    'avocado': 8.5
}
cart ={}  #Empty dictionary to store purchased items and quantities.

total_cost =0 # Variable to store the final total cost.
# Lets the user "add" items by typing their names.

# Loop through the groceries dictionary.
# key = item name, value = price
for key, value in groceries.items():
    print(f'item name: {key}, price:${value}')

# Keeps asking the user for items until they types "checkout".
print()   

while True:
   #Read the item name from the user 
   grocery=input('enter the item want to buy or checkout: ').lower()

   # Stop the loop if the user types checkout.
   if grocery=='checkout':
      break
   
   # Check whether the item exists in the groceries dictionary.
   if grocery in groceries:
       # Ask the user for the quantity.
        quantity = int(input('Enter the number item you buy: ' ))

         # If the item is already in the cart,
        # add the new quantity to the existing quantity.
        if grocery in cart:
           cart[grocery]= cart[grocery] + quantity
         #otherwise, add the item to the cart 
        else:
           cart[grocery]=quantity
        print(f"{grocery} added to cart.")
    # Execute if the item does not exist.
   else:
    print('Invalid item name')


print()
print('\n The Final Bill')
print('--'*20)

# quantity = number purchased
for key, quantity in cart.items():
   price = groceries[key]
   subtotal = price*quantity
   total_cost = total_cost + subtotal

   print(f'{key}, {quantity} * ${price} = ${subtotal}')
print('--'*20)
print(f'{'Total'}= ${total_cost}')
```

# 3)	Build a to-do list manager that

i. 	    Allows users to add tasks with priorities (e.g., "Buy milk - high").

ii.	    Lets them view the current list, delete tasks by number, and mark tasks as complete.

iii. 	Keeps looping until the user types "exit".

iv.  	Shows a summary at the end: number of completed tasks vs pending.

```python
todo_list = []  # each item: [task, priority, completed]

while True:
    choice = input('\n1. Add task 2. view tasks, 3. Delete task 4. complete 5. Exit :')

    # 1. Add task
    if choice == "1":
        task = input("Enter task: ")
        priority = input("Enter priority (high/medium/low): ")
        todo_list.append([task, priority, False])   #todolist = [buy, reding,   ....]
        #print("Task added successfully!")

    # 2. View tasks
    elif choice == "2":
        for i, (task, priority, Done) in enumerate(todo_list, start=1):
            status = 'Done' if Done else 'pending'
            print(f'{i}. {task}- {priority}-{status}')

    #deleting task
    elif choice =='3':
        num = int(input('enter the task number to delete: '))
        if 1<=num<=len(todo_list):
            todo_list.pop(num-1)

    #marking complete
    elif choice =='4':
        num = int(input('task number to complete: '))
        if 1<=num<=len(todo_list):
            todo_list[num-1][2]=True
    
    #Exit
    elif choice=='5':
        break
    else:
        print('Invalid choice')
print(todo_list)

completed = sum(task[2] for task in todo_list)
pending = len(todo_list)-completed

print('\nSummary')
print('completed', completed)
print('pending', pending)
            
```

Build a program that:
•	Remove invalid records where:
o	user is empty OR
o	action is missing (None)
o	Count actions per user (dictionary)
o	Count frequency of each action (dictionary)
o	Cleaned record count
o	User activity summary
o	Most common action


Expected Output
Cleaned Records: 4

    "alice": 2,
    "bob": 2,
    "charlie": 0

```python
logs = [
    {"user": "alice", "action": "login"},
    {"user": "bob", "action": "login"},
    {"user": "alice", "action": "purchase"},
    {"user": "", "action": "login"},
    {"user": "charlie", "action": None},
    {"user": "bob", "action": "logout"}
]


user_activity={}  #stores number of action per users 
cleaning_logs=[]   #record only valid records
action_counts={}   #stores frequency of each action   

    
for log in logs:
    user=log['user']
    action =log['action']

    if user != '' and user not in user_activity:
        user_activity[user]=0
    
    if user == '' or action is None:
        continue
 
    cleaning_logs.append(log)   #only take the valid records
   
    if user in user_activity:
        user_activity[user]= user_activity[user] +1
    else:
        user_activity[user]=1


    if action in action_counts:                     #counting the each action
         action_counts[action] = action_counts[action]+1
    else:
        action_counts[action]=1

Most_common_action= max(action_counts, key=action_counts.get)

print('user_activity :', user_activity)


print('cleaning_records:', len(cleaning_logs))
#print(cleaning_logs)


print('number of action records: ', action_counts)

print('most common action: ', Most_common_action)

```
